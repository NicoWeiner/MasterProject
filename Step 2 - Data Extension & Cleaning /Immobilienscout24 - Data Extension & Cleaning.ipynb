### 3.4.  Calculate timetosell

```python
immomerged2_1 = immomerged3
```

Following creates a new column 'timetosell' that shows that a listing has sold between the first and following scraping dates. 

The logic is that a house is assumed to be sold when it does not appear in the newly scraped dataset. If a value in 'ScrapingDate2', 'ScrapingDate3' etc is hence NaT, the column 'timetosell' shows how many days a listing was on the market since the first scraping date at 21st of April, 2020 until the first follow up ScrapingDate that has a NaT value. 

This solution is limited in validity as it is unknown if a listing was really sold at the day when it was removed from Immobilienscout24 or for how long it already had been listed before the first scraping date. Still, it allows to generate an estimated figure that is better than having no insights about listing periods at all.  

```python
# create dummy column where the timetosell is aggregated in:
immomerged2_1['timetosell'] =  'NaN'

# create dummy columns with number of days between scraping dates. 
# If a row has a date, it is still on market. 
# If a row does not have a date anymore with every new scraped data, it is assumed to be sold. 
immomerged2_1['timetosell2'] = np.where(immomerged2_1['ScrapingDate2']=='2020-04-25', 'NaN', '4')
immomerged2_1['timetosell1'] = np.where(immomerged2_1['ScrapingDate3']=='2020-04-29', 'NaN', '8')
immomerged2_1['timetosell'] = np.where(immomerged2_1['ScrapingDate4']=='2020-05-03', 'NaN', '12')



# When a fourth dataset with ScrapingDate4 is available, also run following line:  
# immomerged3['timetosell'] = np.where(immomerged3['ScrapingDate4']=='2020-05-03', 'NaN', '12')

# When a fifth dataset with ScrapingDate5 is available, also run following line:  
# immomerged4['timetosell'] = np.where(immomerged4['ScrapingDate5']=='2020-05-07', 'NaN', '16')


# the above code takes 8 days for all houses. As some are already sold after 4 days, 
# I adapt those in the final timetosell column. 
immomerged2_1.loc[immomerged2_1["timetosell1"].str.contains("8"), "timetosell"] = "8"
immomerged2_1.loc[immomerged2_1["timetosell2"].str.contains("4"), "timetosell"] = "4"



# now drop the dummy columns again
immomerged2_1 = immomerged2_1.drop(['timetosell1', 'timetosell2'], axis=1)

#as a result, we see a timetosell column with amount of days needed per listing until being sold
immomerged2_1.head(30)
```

```python
immodf2.count()
```

immodf2 now contains all usable listing IDs of Immobilienscout24. In the next step, I convert all other columns to the particular dtype that their values are in. 

-> as the data is in German format, for some columns, "," has to be replaced with "."

```python

immodf2['Kaufpreis'] = immodf2['Kaufpreis'].str.replace('€', '')
immodf2['Kaufpreis'] = immodf2['Kaufpreis'].str.replace('.', '')
immodf2['Kaufpreis'] = immodf2['Kaufpreis'].str.replace(',', '.')
immodf2['Anzahl_zimmer'] = immodf2['Anzahl_zimmer'].str.replace(',', '.')
immodf2['Baujahr'] = immodf2['Baujahr'].str.replace('unbekannt', 'NaN')
immodf2['Wohnflaeche (in m²)'] = immodf2['Wohnflaeche (in m²)'].str.replace('.', '')
immodf2['Wohnflaeche (in m²)'] = immodf2['Wohnflaeche (in m²)'].str.replace(',', '.')
immodf2['Grundstueck (in m²)'] = immodf2['Grundstueck (in m²)'].str.replace('.', '')
immodf2['Grundstueck (in m²)'] = immodf2['Grundstueck (in m²)'].str.replace(',', '.')

immodf2.head(20)
```
